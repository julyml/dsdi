```python
import pandas as pd


df_xml_rpc_same_computer = pd.read_csv('analise_xml_rpc_same_computer.csv')
df_xml_rpc_same_computer_media_por_experimento = df_xml_rpc_same_computer.groupby(['experimento'])['duracao'].mean()
df_xml_rpc_same_computer_desvio_padrao_por_experimento = df_xml_rpc_same_computer.groupby(['experimento'])['duracao'].std()

print(df_xml_rpc_same_computer_media_por_experimento)
print(df_xml_rpc_same_computer_desvio_padrao_por_experimento)
```

```python
df_grpc_same_computer = pd.read_csv('analise_grpc_same_computer.csv')
df_grpc_same_computer_media_por_experimento = df_grpc_same_computer.groupby(['experimento'])['duracao'].mean()
df_grpc_same_computer_desvio_padrao_por_experimento = df_grpc_same_computer.groupby(['experimento'])['duracao'].std()

print(df_grpc_same_computer_media_por_experimento)
print(df_grpc_same_computer_desvio_padrao_por_experimento)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Dados dos experimentos com gRPC
grpc_experimentos = ['Void', 'Long', 'Oito_long', 'String']
grpc_medias = [0.0019, 0.0007, 0.0004, 0.0005]
grpc_desvio_padrao = [0.0030, 0.0004, 0.0005, 0.0004]

# Dados dos experimentos com XML-RPC
xmlrpc_experimentos = ['Void', 'Long', 'Oito_long', 'String']
xmlrpc_medias = [2.0335, 2.0396, 2.0330, 2.0337]
xmlrpc_desvio_padrao = [0.0100, 0.0057, 0.0061, 0.0092]

# Configuração do estilo dos gráficos
sns.set(style="whitegrid")

# Gráfico de barras para as médias
plt.figure(figsize=(10, 6))
sns.barplot(x=grpc_experimentos, y=grpc_medias, color='blue', alpha=0.5, label='gRPC')
sns.barplot(x=xmlrpc_experimentos, y=xmlrpc_medias, color='red', alpha=0.5, label='XML-RPC')
plt.title("Médias dos Experimentos")
plt.xlabel("Experimento")
plt.ylabel("Média")
plt.legend()
plt.show()

# Gráfico de barras para os desvios padrão
plt.figure(figsize=(10, 6))
sns.barplot(x=grpc_experimentos, y=grpc_desvio_padrao, color='blue', alpha=0.5, label='gRPC')
sns.barplot(x=xmlrpc_experimentos, y=xmlrpc_desvio_padrao, color='red', alpha=0.5, label='XML-RPC')
plt.title("Desvios Padrão dos Experimentos")
plt.xlabel("Experimento")
plt.ylabel("Desvio Padrão")
plt.legend()
plt.show()
```
